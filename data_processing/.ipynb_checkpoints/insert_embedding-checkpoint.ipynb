```python
import pandas as pd
```

```python
df = pd.read_csv("../data/movies_cleaned.csv")
df.head(5)
```

```python
# get number of rows
print(df.shape[0])
```

```python
import ollama


def get_embedding(text, max_chars=500):
    text = text if isinstance(text, str) else str(text)
    if len(text) > max_chars:
        text = text[:max_chars]
    return ollama.embeddings(
        model="mxbai-embed-large",
        prompt=text,
    )["embedding"]
```

```python
from qdrant_client import QdrantClient

client = QdrantClient(url="http://localhost:6333")
```

```python
from tqdm import tqdm
from qdrant_client.models import PointStruct
import time

start = 0


for index, row in tqdm(df[start:].iterrows(), total=df[start:].shape[0]):

    row["overview"] = row["overview"].encode("utf-8").decode("utf-8")

    payload = {
        "title": row["title"],
        "overview": row["overview"],
        "release_date": row["release_date"],
        "revenue": row["revenue"],
        "runtime": row["runtime"],
        "budget": row["budget"],
        "genre": row["genre"],
        "production_companies": row["production_companies"],
        "vote_average": row["vote_average"],
        "vote_count": row["vote_count"],
    }

    embedding = get_embedding(row["overview"])
    client.upsert(
        collection_name="movies",
        wait=True,
        points=[PointStruct(id=index, payload=payload, vector=embedding)],
    )
```
